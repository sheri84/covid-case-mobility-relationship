```python
import sys
import pandas as pd
import pymongo
import json
import os
```

```python
   ##connect to mongodb, select database and collection
mng_client = pymongo.MongoClient('localhost', 27017)
mng_db = mng_client['covid_project'] # Replace mongo db name
collection_name = 'mobility' # Replace mongo db collection name
db_cm = mng_db[collection_name]
```

```python
   ##define the file path and load with pandas
filepath="Global_Mobility_Report.csv"
data = pd.read_csv(filepath)
```

```python
data_sub = data[data["country_region"]=="Australia"]
```

```python
  ##CONVERTING DATAFRAME INTO JSON FORMAT
data_json = json.loads(data_sub.to_json(orient='records'))

```

```python
db_cm.remove()
db_cm.insert(data_json)
```
